```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Defines the path where the image folders of different types of bacteria are located.
new_image_dir = r'C:\Users\bruna\OneDrive\Microbialdataset\Culture_dataset'
```

```python
# Step 2: List the images and count
bacteria_counts = {}

# List the types of bacteria (folder names)
for folder in os.listdir(new_image_dir):
    folder_path = os.path.join(new_image_dir, folder)
    if os.path.isdir(folder_path):
        # Count the number of image files in the folder
        image_count = len([file for file in os.listdir(folder_path) if file.endswith(('.png', '.jpg', '.jpeg'))])
        # Store the count in the dictionary
        bacteria_counts[folder] = image_count
```

```python
# Convert the dictionary into a DataFrame
bacteria_df = pd.DataFrame(list(bacteria_counts.items()), columns=['Bacteria_Type', 'Image_Count'])
```

```python
# Bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Bacteria_Type', y='Image_Count', data=bacteria_df)
plt.xlabel('Type of bacteria')
plt.ylabel('Number of images')
plt.xticks(rotation=45)
plt.show()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(bacteria_df['Image_Count'], labels=bacteria_df['Bacteria_Type'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Para garantir que o gráfico de pizza seja um círculo
plt.show()
```
